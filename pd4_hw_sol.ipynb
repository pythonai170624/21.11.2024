```python
import numpy as np
import pandas as pd
mobile = pd.read_csv(r"mobile.csv")

comp_names =  {'050': 'Pelephone', '051': 'Golan', 
                   '052': 'Cellcum', '053': 'Motorolla',
                   '054': 'Partner'}

# 1
mobile['First name']  = mobile['Customer Name'].apply(lambda name: name.split()[0])
print(mobile['First name'].head(3))
print()

# 2
def company_name(num):
    return comp_names.get(str(num[:3]), 'NA')
    
mobile['Company'] = mobile['Phone Number'].apply(lambda num: comp_names.get(num[:3], 'NA'))
# 3a
print(mobile['Company'].value_counts())
# 3b
print()
print(mobile['Company'].unique())
# 3c
print(mobile['Company'].nunique())

# 4
# 4a
mobile['type'] = mobile.apply(lambda row: "Talker" if row['Talked'] > row['Internet'] 
                              else "Surfer", axis=1)
print(mobile.head(3))
# 4a
mobile['type2'] = np.vectorize(lambda talked, internet: "Talker" if talked > internet 
                               else "Surfer")(mobile['Talked'],mobile['Internet'])
print(mobile.head(3))

# 5
print()
print(mobile.sort_values(['Talked'], ascending=True).head(5))

# 6
print()
print(mobile.sort_values(['Talked', 'Internet'], ascending=[True, False]).head(5))

# 7
print()
print(mobile.nlargest(5, 'Talked'))

# 8
print()
print(mobile.nsmallest(3, 'Internet'))
```
